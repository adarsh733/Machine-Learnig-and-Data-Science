```python
fig=plt.figure()
axes=fig.add_axes([0.1,0.1,1,1])

axes.plot(x,x**2,label='X Squared',color='blue',lw=2,ls='-.',marker="o")
axes.plot(x,x**3,label='X Cubed',color='black',lw=2,ls='-',marker="o",markersize=10,markerfacecolor="yellow",markeredgecolor='green')
 
axes.legend(loc=0)

```
