# Grids

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
iris=sns.load_dataset('iris')
```

```python
iris.head()
```

```python
iris['species'].unique()
```

```python
g=sns.PairGrid(iris)
g.map_diag(plt.hist)
g.map_upper(plt.scatter)
g.map_lower(sns.kdeplot)

```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',markers=['o','v'],scatter={'s',100})
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,hue='sex',col='day',aspect=0.6,size=8)
```

# Colors and Style

```python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline
tips=sns.load_dataset('tips')
```

```python
sns.set_style(style='darkgrid')
sns.countplot(x='sex',data=tips)
sns.despine(left=True)
```

```python
plt.figure(figsize=(12,12))

sns.countplot(x='sex',data=tips,palette='deep')
```

```python
sns.lmplot(x='total_bill',y='tip',data=tips,palette='deep',aspect=10,size=10)
```

```python
sns.set_context(context='poster',font_scale=3)
sns.countplot(x='sex',data=tips,palette='deep')
```
